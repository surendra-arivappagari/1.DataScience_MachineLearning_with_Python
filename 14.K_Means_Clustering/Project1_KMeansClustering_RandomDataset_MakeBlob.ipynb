### A). Project Overview

* K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. 
* In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. 
* Then, the algorithm iterates through two steps: Reassign data points to the cluster whose centroid is closest. Calculate new centroid of each cluster. 
* These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

### C). Create the data for our machine learning project

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
#n_samples=no..of rows, n_features=no..of columns, centers=no..of clusters

data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,random_state=101)
```

```python
type(data)
```

#### D). visualize the data

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

### E). Creating the Clusters

```python
#import Kmeans from sklearn.cluster

from sklearn.cluster import KMeans
```

```python
#create an instance with 4 clusters

kmeans = KMeans(n_clusters=4)
```

```python
#train the data

kmeans.fit(data[0])
```

```python
#check the centers of each cluster 

kmeans.cluster_centers_
```

```python
#check the labels

kmeans.labels_
```

```python
#compare the KMeans algorithem and original data


fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title("K Means:")
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title("Original:")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

#### Both graphs (KMeans and original )are matching.
