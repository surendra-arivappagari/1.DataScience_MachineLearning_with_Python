# Pandas Tutorials

> - Pandas is used for Processing (load, manipulate, prepare, model, and analyze) the given data. <br>
> - Pandas is built on top of the Numpy package so Numpy is required to work with Pandas <br>
> - Pandas has 2 data structures for processing the data.
> > 1. Series    --> is a one-dimensional array that is capable of storing various data types.
> > 2. DataFrame --> is a two-dimensional array with labeled axes (rows and columns)

# Import required modules

# A) Series in Pandas

```python
# Series a one-dimensional array that is capable of storing various data types
# The row labels of series are called the index.
# Series is capable of holding Integers, strings, floating point numbers, Python objects.
# It should not contains multiple columns.

print("Series Example")
s = pd.Series(np.arange(15,20))
s
```

### A1) Series from ndarray

```python
print("Series From ndarray. \n\nSeries:")

s = pd.Series(np.random.randn(5), index=['a','b','c','d','e'])
print(s)
```

### A2) Series from Dictionary

```python
print("Series From Dictionary. \n\nSeries:")

dict1 = {'p':111, 'q':222, 'r':333, 's':np.NaN, 't':555}
s = pd.Series(dict1)
print(s)
```

### A3) Series from Scalar value

```python
print("Series From Scalar value. Same value will be repeated with index length. \n\nSeries:")

s = pd.Series(125, index=['i','j','k','l'])
print(s)
```

### A4) Series functionalities

```python
print("Series will work like ndarray. Slice operations \n\nSeries:")

dict1 = {'p':111, 'q':222, 'r':333, 's':np.NaN, 't':555}
s = pd.Series(dict1)

print(s)
```

```python
print("Slice:\ns[1]:",s[1])
print("s['r']:",s['r'])

print("\n##############################")
print("Filters:s[s > 200]:\n", s[s > 200])

print("\n##############################")
print("Select Multiple indexes:s[0,2,4]:\n", s[[0,2,4]])

print("\n##############################")
print("Check DType:", s.dtype)

```

```python
print("\n##############################")
print("Sum 2 Series: s + s :\n", s + s)

print("\n##############################")
print("Multiply by 5 : s*5 :\n", s*5)
```

### B9) Missing data

> ###### None --> it is python default value for missing the data
> ###### NaN  --> This is from Numpy package(np.nan), widely used in numpy arrays , Pandas Series and Dataframes
>> ###### 1.isnull() --> Generate a boolean mask indicating missing values
>> ###### 2.notnull() --> Generate a boolean mask indicating proper values(it is opposite to isnull())
>> ###### 3.fillna() -->we can replace NA with a scalar value
>> ###### 4.dropna() --> used to drop the row or column if it have missing data 


```python
#Creating sample DF with null values.
print("Example for DataFrame having NaN's.")

df = pd.DataFrame(np.random.rand(5,4), index =[2010, 2013, 2015, 2016, 2020], columns = 'A B C D'.split())
df = df.reindex(np.arange(2010,2021))
df['C'] = np.random.rand(11)
df.loc[2014] = np.nan
df.loc[2019] = np.nan
df['E'] = np.random.rand(11)
df['F'] = np.nan
df
```

#### B9a) isnull()

```python
print("print Boolean array with isnull() method.")
df.isnull()
```

#### B9b) notnull()

```python
print("print Boolean array with notnull() method. This is completely opposite to isnull()")
df.notnull()
```

#### B9c) fillna(value, method=[ffill,bfill], axis=[0,1])

```python
print("Example1: for fillna() method. Fill all missing values with 0 in entire DF.")
df.fillna(0)
```

```python
print("Example2: for fillna() method. Fill all missing values with some string in entire DF.")
df.fillna("Missing")
```

```python
#ffill--> will copy the data in forward(Vertically) direction in all columns.If no data in entore column remainn same.

print("Example3: for fillna() method. Forward(Vertically) filling with data in all columns.")
df.fillna(method = 'ffill')
```

```python
#ffill--> will copy the data in forward(Horizontally) direction in all columns.If no data in entore column remainn same.

print("Example3b: for fillna() method. Forward(Horizontally) filling with data in all columns.")
df.fillna(method = 'ffill', axis=1)
```

```python
#ffill--> will copy the data in backward(Horizontally) direction in all columns.If no data in entore column remainn same.

print("Example4: for fillna() method. Backward(Horizontally) filling with data in all columns.")
df.fillna(method = 'bfill')
```

```python
#ffill--> will copy the data in backward(Horizontally) direction in all columns.If no data in entore column remainn same.

print("Example4: for fillna() method. Backward(Horizontally) filling with data in all columns.")
df.fillna(method = 'bfill', axis=1)
```

### B11) Combine multiple dataframes
>> #### B11a) merge()  --> combining data on common columns or indices. [We specify the join columns here]
>> #### B11b) join()   --> combining data on an index. [We wont specify anything here on what bases it should join], it is just like Left join in SQL.
>> #### B11c) concat() --> combining DataFrames across rows or columns

```python
print("Sample DF's used in this practice.")

#df1
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3','A4', 'A5'],
                        'B': ['B0', 'B1', 'B2', 'B3', 'B4', 'B999'],
                        'C': ['C0', 'C1', 'C2', 'C3', 'C4', 'C5'],
                        'D': ['D0', 'D1', 'D2', 'D3', 'D4', 'D5']},
                        index=[0, 1, 2, 3, 4, 5])

#df2
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7'],
                        'E': ['E4', 'E5', 'E6', 'E7']},
                         index=[4, 5, 6, 7]) 

#df3
df3 = pd.DataFrame({'A': ['A7', 'A8', 'A9', 'A10', 'A11'],
                        'B': ['B7', 'B8', 'B9', 'B10', 'B11'],
                        'C': ['C7', 'C8', 'C9', 'C10', 'C11'],
                        'D': ['D7', 'D8', 'D9', 'D10', 'D11']},
                        index=[7, 8, 9, 10, 11])

print("\n DF1:\n", df1)
print("\n DF2:\n", df2)
print("\n DF3:\n", df3)
```

#### B11a) Merge()

```python
#Merge(left_df, right_df, how=[inner, left, right, outer], left_on=[key in left_df], right_on=[keys in right_df])
# Merge is the common method used in many cases. Just like Joins in SQL.

print("Ex1: Inner Join using Merge.Matching values in all columns.")
#df1_merge_df2 = pd.merge(df1, df2) --> by default merge is inner merge
df1_inner_df2 = pd.merge(df1, df2, how='inner')
df1_inner_df2
```

```python
print("Ex2: Inner Join using Merge.Matching values in A, C, D columns.")

df1_inner_df2 = pd.merge(df1, df2, how='inner', on=['A', 'C', 'D'])
df1_inner_df2
```

```python
print("Ex3: Left Join using Merge. All rows[6] from left, matching rows[2] from Right DF. Output rows=6.")

df1_left_df2 = pd.merge(df1, df2, how='left', on=['A', 'D'])
df1_left_df2
```

```python
print("Ex4: Right Join using Merge. All rows[4] from right, matching rows[2] from left DF. Output rows=4.")

df1_right_df2 = pd.merge(df1, df2, how='right', on=['A', 'D'])
df1_right_df2
```

```python
print("Ex5: Outer Join using Merge. leftDF[6]-matching[2] + matching[2] + rightDF[4]-matching[2]. Output rows =8.")

df1_outer_df2 = pd.merge(df1, df2, how='outer', on=['A', 'D'])
df1_outer_df2
```

#### B11b) Join()

```python
print("Sample DF's for join() method.\n")

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

df2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
print("\nDF1: \n", df1)
print("\nDF2: \n", df2)
```

```python
#joins are bydefault acts like left joins in SQL.

print("Ex1: Join() example. Here we wont specify on what bases it should join, it will consider indexes for joining.")
df1_join_df2 = df1.join(df2)
df1_join_df2
```

```python
print("Ex2: Join() with outer join. We wont specify any matching criteria.")
df1_join_df2 = df1.join(df2, how='outer')
df1_join_df2
```

#### B11c) concat()

```python
# concat() is used for combining Data Frames across rows or columns.
# We can concat data vertically or horizontally on 1 Df to other df.

print("Sample DF's used in this practice.")

#df1
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3','A4', 'A5'],
                        'B': ['B0', 'B1', 'B2', 'B3', 'B4', 'B999'],
                        'C': ['C0', 'C1', 'C2', 'C3', 'C4', 'C5'],
                        'D': ['D0', 'D1', 'D2', 'D3', 'D4', 'D5']},
                        index=[0, 1, 2, 3, 4, 5])

#df2
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7'],
                        'E': ['E4', 'E5', 'E6', 'E7']},
                         index=[4, 5, 6, 7]) 

#df3
df3 = pd.DataFrame({'A': ['A7', 'A8', 'A9', 'A10', 'A11'],
                        'B': ['B7', 'B8', 'B9', 'B10', 'B11'],
                        'C': ['C7', 'C8', 'C9', 'C10', 'C11'],
                        'D': ['D7', 'D8', 'D9', 'D10', 'D11']},
                        index=[7, 8, 9, 10, 11])

print("\n DF1:\n", df1)
print("\n DF2:\n", df2)
print("\n DF3:\n", df3)
```

```python
#If we concat 2 dataframes basedon indexes, their #of columns should match.
print("Ex1: Concat DataFrames vertically. \nSo #of Columns should match. If columns count not matches, then it will give warning.\n")

#pd.concat([df1, df2, df3], axis=0)--> both are same.
df1_df2_df3_ver_concat = pd.concat([df1, df2, df3])
df1_df2_df3_ver_concat
```

```python
#concatinating 2 dataframes basedon columns.(Horizontaly.)
print("Ex1: Concat DataFrames Horizontally.")

df1_df2_df3_hor_concat = pd.concat([df1, df2, df3], axis=1)
df1_df2_df3_hor_concat
```

### B12) Apply lambda functions and user defined functions to columns of DataFrame 

```python
print("Sample DF.\n")
dict1 = {"District": ['Anantapur', 'Anantapur', 'Anantapur', 'Kadapa', 'Kadapa', 'Kadapa'],
        "City": ['Hindupur', 'Penukonda', 'Roddam', 'RKV', 'Pulivendula', 'Vempalli'],
        "Population": [30, 18, 15, 22, 29, 14],
        "Tax": [30.45, 18.45, 15.45, 22.65, 29.65, 14.65]}

df = pd.DataFrame(dict1)
df
```
