### B). Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### C). Create a Dataframe

```python
#data = actual data, feature_names = list of columns

df_cancer = pd.DataFrame(cancer['data'], columns= cancer['feature_names'])
df_cancer.head()
```

```python
#nulls, column datatype information
df_cancer.info()
```

```python
#statistical information for numerical columns

df_cancer.describe()
```

```python
#check the target in cancer dataset

print("Target_names:", cancer['target_names'], "\n\n Target:")

df_target = cancer['target']
df_target
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
#X = all nemerical columns except target. bydefault in sklearn datasets target will be different column.
#y = target

X = df_cancer
y = np.ravel(df_target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```
