#### 1A) Distplot():

```python
print("EX1: The default plot kind is a histogram:")

sns.distplot(tips['total_bill'])
```

```python
print("EX2: Remove KDE(Kernel density estimation) layer.")

sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
#Data 
penguins = sns.load_dataset("penguins")
penguins.head()
```

#### 1B) Jointplot():

```python
#with this we can generate these plots. kind{ scatter | kde |  hex | reg | resid }
print("EX1: Jointplot() we can generate scatter plot.")

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter')
```

```python
print("EX2: using Jointplot() we can generate hex plot.")

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='hex')
```

```python
print("EX3: using Jointplot() we can generate reg plot. Used to add a linear regression fit.")

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='reg')
```

```python
print("EX4: using Jointplot() we can generate kde plot.")

sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde")
```

```python
print("EX5: using Jointplot() we can Comibne multiple graphs.")

g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
```
