## 1) Distribution Plots:

-  These plots useful to check how variables(columns) are distributed.

#### 1A) Distplot():

- `Distplot()` will show a histogram with a line on it. <br>
- `Distplot()` plots a univariate distribution of observations. The distplot() function combines the matplotlib hist function with the seaborn kdeplot() and rugplot() functions

```python
print("1A-EX1: The default plot kind is a histogram along with rugplot.:")

sns.distplot(tips['total_bill'])
```

```python
print("1A-EX2: Remove KDE(Kernel density estimation) layer.")

sns.distplot(tips['total_bill'],kde=False)
```

```python
print("1A-EX3: Remove Histograms and show only KDE.")

sns.distplot(tips['total_bill'],kde=True, hist=False)
```

```python
print("1A-EX4: Draw distplot in vertically.")

sns.distplot(tips['total_bill'],vertical=True)
```

```python
print("1A-EX5: Change the color.")

sns.distplot(tips['total_bill'],color='green')
```

```python
print("1A-EX6: Distplot() with regplot.")

sns.distplot(tips['total_bill'], rug=True)
```

#### 1B) Kdeplot(): Kernel Distribution Estimation

- `Kdeplot()` is used to plot the data against a single/univariate variable. It represents the probability distribution of the data values as the area under the plotted curve. <br>
- `Kdeplot()` used to plot the data against multiple data variables/bivariate(2) variables to check the probability distribution of one with respect to the other values.

```python
print("1B-EX1: Sample Kdeplot()- Area under the curve- Probability Distribution.")

sns.kdeplot(tips['total_bill'])
```

```python
print("1B-EX2: Sample 2 Kdeplots()-with some parameters.")

sns.kdeplot(tips['total_bill'], color='green', shade=True)
sns.kdeplot(tips['tip'], color='blue', shade=True)
```

```python
print("1B-EX3: Kdeplot() with bivariant Example. X,Y(Total_bill vs tip) ")

sns.kdeplot(tips['total_bill'], tips['tip'], color='blue', shade=True)
```

```python
print("1B-EX4: Sample Kdeplot()- vertical.")

sns.kdeplot(tips['total_bill'], vertical=True)
```

```python
print("1B-EX5: Kdeplot() with bivariant Example. X,Y(Total_bill vs tip) ")

sns.kdeplot(tips['total_bill'], tips['tip'], color='blue', shade=True, cmap='Purples_d', cbar=True)
```

#### 1C) Rugplot(): 

- `Rugplot()` it will places a horizontally dash with each occurrence of an item in a dataset. <br>
- Areas where there is great occurrence of an item see a greater density of these dashes. viceverse.

```python
print("1C-EX1: Combine Rug with KDE to see more meaningful.")

sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

#### 1D) Jointplot(): 

- `Jointplot()` will displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.  <br>
- we can generate these plots. kind{ scatter | kde |  hex | reg | resid }

```python
print("1D-EX1: Jointplot() we can generate scatter plot.")

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter')
```

```python
print("1D-EX2: using Jointplot() we can generate hex plot.")

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='hex')
```

```python
print("1D-EX3: using Jointplot() we can generate reg plot. Used to add a linear regression fit.")

sns.jointplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='reg')
```

```python
print("1D-EX4: using Jointplot() we can generate kde plot.")

sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde")
```

```python
print("1D-EX5: using Jointplot() we can Comibne multiple graphs.")

g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
```

#### 1E) Pairplot(): 

- `Pairplot()` will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).   <br>
- `Pairplot()` will generate n X n size of numerical columns subplots.

```python
print("1E-EX1: Pairplot() sample. In Penguins DF we have 4 numerical columns.")

sns.pairplot(penguins)
```

```python
#Hue will give more insightful with categorical columns.

print("1E-EX2: Pairplot() sample.Hue for categorical columns.")

sns.pairplot(penguins, hue='species')
```

```python
print("1E-EX3: Pairplot() sample. Customized diagonal with histplot. ")

sns.pairplot(penguins, hue='species', diag_kind="hist")
```

```python
print("1E-EX4: Pairplot() sample. Markers will be assigned to each category.")

sns.pairplot(penguins, hue='species', markers=['+', '*', '^'])
```

#### 2A) Barplot():
- `Barplot()` is used to get aggregate data off a categorical feature in your data. <br>
- `Barplot()` will represents an estimate of central tendency for a numeric variable with the height of each rectangle.

```python
print("2A-EX1: Draw vertical barplots grouped by a categorical variable: Day vs Total_Bill.")

sns.barplot(data=tips, x='day', y='total_bill')
```

```python
print("2A-EX2: Draw vertical barplots with nested grouped by 2 variable: Day vs Total_Bill for Gender.")

sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
print("2A-EX3: Horizontal bars:")

sns.barplot(data=tips, x='total_bill', y='day')
```

```python
print("2A-EX4: We can control the order of each bar.")

sns.barplot(data=tips, x='day', y='total_bill', order=['Sat', 'Sun', 'Thur', 'Fri'])
```

#### 2B) Countplot():
- `Countplot()` Show the counts of observations in each categorical bin using bars. <br>

```python
print("2B-EX1: Titanic Dataset for countplot().")

sns.countplot(data=titanic, x='class')
```

```python
print("2B-EX2: Show value counts for Class with Who.")

sns.countplot(data=titanic, x='class', hue='who')
```

#### 2C) Boxplot():
- `Boxplot()` shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. <br>


```python
print("2C-EX1: single horizontal boxplot:")

sns.boxplot(x=tips['total_bill'])
```

```python
print("2C-EX2: vertical boxplot grouped by Day variable.")

sns.boxplot(data=tips, x='day', y='total_bill', palette="coolwarm")
```

```python
print("2C-EX3: Boxplot with nested grouping by two categorical(Day, Smoker) variables:")

sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker', palette="coolwarm")
```

#### 2D) Violinplot():
- `Violinplot()`These similar to Boxplots. These shows the distribution of quantitative data across several levels of one (or more) categorical variables. <br>
- `Violinplot()`uses features of kernel density estimation of the underlying distribution.

```python
print("2D-EX1: single horizontal violinplot:")

sns.violinplot(data=tips, x='total_bill')
```

```python
print("2D-EX2: vertical violinplot grouped by a categorical (Day) variable:")

sns.violinplot(data=tips, x='day', y='total_bill',palette="coolwarm")
```

```python
print("2D-EX3: Violinplot with nested grouping by two categorical(Day, Smoker) variables:")

sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker', palette="coolwarm")
```

```python
print("2D-EX4: split violins to compare the across the hue variable. Here Hue should have only 2 distinct values.")

sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker', split=True, palette="coolwarm")
```

#### 2E) Stripplot():
- `Stripplot()` will draw a scatterplot where one variable is categorical. <br>
- `Stripplot()` can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
print("2E-EX1: single horizontal strip plot:")

sns.stripplot(data=tips, x='total_bill')
```

```python
print("2E-EX2: single horizontal strip plot with Jitter for making distribution smaller.")

sns.stripplot(data=tips, x='total_bill', jitter=0)
```

```python
print("2E-EX3: strips by a categorical (Day) variable:")

sns.stripplot(data=tips, x='day', y='total_bill', jitter=0)
```

```python
print("2E-EX4: strips by 2 categorical (Day, Gender) variable. Jitter with 0.08")

sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', jitter=0.08)
```

#### 2F) Swarmplot():
- `Swarmplot()` is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. <br>
- `Swarmplot()` will gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them.

```python
print("2F-EX1: single horizontal swarm plot:")

sns.swarmplot(data=tips, x='total_bill')
```

```python
print("2F-EX2: swarms by a categorical(Day) variable:")

sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
print("2F-EX3: Color the points using a second categorical(Gender) variable:")

sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
print("2F-EX4: Combine swarms of observations on top of a box plot:")

sns.boxplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
print("2F-EX5: Combine swarms of observations on top of a violinplot:")

sns.violinplot(data=tips, x='day', y='total_bill', inner=None)
sns.swarmplot(data=tips, x='day', y='total_bill', color="white")
```

## 3) Matrix Plots

- Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data.
> 3A) `Heatmap()`: <br>
> 3B) `Clustermap()`:  <br>


#### 3A) Heatmap():
- `Heatmap()` In order for a heatmap to work properly, our data should already be in a matrix form, this function just colors it the given matrix.<br>
- `Heatmap()` cannot be applied to any DF directly without converting the DF into Matrix form(Pivoting table.)

```python
print("3A-EX1: Try with Correlated matrix for tips DF.")

print("\nCorrelated Matrix of Tips DF:")
print(tips.corr())

print("\n\nApplying Heatmap() to above DF.")
sns.heatmap(tips.corr())
```

```python
print("3A-EX2:Correlated matrix for tips DF with Annotations.")

sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

```python
print("3A-EX3A: HeatMap for Flight data.Cehck how data is there in Flight DF.")
print(flights.head())

print("\nApply Pivot to above data. By Months as Index, Year as columns, value as Passengers.")
flights.pivot_table(index='month', columns='year', values='passengers')


#sns.heatmap(tips.corr(), cmap='coolwarm', annot=True)
```

```python
print("3A-EX3B: HeatMap for Flight data, After converting to Pivot table.")

pivot_flights=flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(pivot_flights)
```

```python
print("3A-EX4: HeatMap Annotate each cell with the numeric value using integer formatting:")

pivot_flights=flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(pivot_flights, annot=True, fmt='d')
```

```python
print("3A-EX5: HeatMap Lines between each cell and color.")

pivot_flights=flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(pivot_flights, linewidths=0.5, linecolor='green')
```

#### 3B) Clustermap():
- `Clustermap()` uses hierarchal clustering to produce a clustered version of the heatmap. <br>
- `Clustermap()` will provide the data by grouping the similar in values. Order of the DF will change.

```python
print("3B-EX1: Clustermap for Pivoted flights DF.")

sns.clustermap(pivot_flights)
```

```python
print("3B-EX2: Use a different colormap and adjust the limits of the color range:")

sns.clustermap(pivot_flights, cmap='mako')
```

```python
print("3B-EX3: ClusterMap() with More options to get information a little clearer like normalization")

sns.clustermap(pivot_flights, cmap='coolwarm',standard_scale=1)
```

## 4) Grid Plots

- Grids are allow us to map the different plottypes to grid rows and columns. <br>
- Gridplots helps us to customize the creation of  similar character/separated plots.
> 4A) `Pairgrid()`: `Pairplot()` are looks similar. <br>
> 4B) `Facetgrid()`:  <br>
> 4C) `Jointgrid()`: `Jointplot()` are looks similar. <br>


#### 4A) PairGrid() look similar to Pairplot():
- `PairGrid()` is a subplot grid for plotting pairwise relationships in a dataset.<br>
- `PairGrid()` will looks like pairplot() template without graphs.

```python
print("4A-EX1: We know in Penguins DF we have 4 numerical columns. Plot a grid for 4X4 template.")

sns.PairGrid(penguins)
```

```python
print("4A-EX2: Fill above 4X4 empty grids exactly same as pairplots using Map with scatter plots.")

grid_tem = sns.PairGrid(penguins)
grid_tem.map(sns.scatterplot)
```

```python
print("4A-EX3: Customize the diagonals with KDE, remining as scatter plots. ")

grid_tem = sns.PairGrid(penguins)
grid_tem.map_diag(sns.kdeplot)
grid_tem.map_offdiag(sns.scatterplot)
```

```python
print("4A-EX4: Customize the diagonals with KDE, Above, below for Diagonal. ")

# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
print("4A-EX5: Pairplot() we alredy seen it. It is very much similar to pairgrid.")

sns.pairplot(penguins, hue='species')
```
