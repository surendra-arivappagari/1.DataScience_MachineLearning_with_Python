# Pandas Build-in Data Visualization:

- In `Pandas` we can able to generate the visualizations wihtout using `Matplotlib`, `seaborn` libraries. <br>
- In the backend `Pandas` uses `Matplotlib` for data visualizations. <br>
- If we want to visualize some data in quick way(without worring too much about parameters) we use this Pandas Build-in stuff for visualization. <br>
- We are using below Plot() types. **Note:** df.plot(kind='hist') is same as df.plot.hist()
> 1) `df.plt.line()`: Line plot (default) <br>
> 2) `df.plt.bar()`: Vertical bar plot <br>
> 3) `df.plt.barh()`: Horizontal bar plot <br>
> 4) `df.plt.hist()`: Histogram <br>
> 5) `df.plt.box()`: Boxplot <br>
> 6) `df.plt.kde()`: Kernel Density Estimation plot <br>
> 7) `df.plt.density()`: it is same as ‘kde’ plot <br>
> 8) `df.plt.area()`: Area plot <br>
> 9) `df.plt.pie()`: Pie plot <br>
> 10) `df.plt.scatter()`: Scatter plot <br>
> 11) `df.plt.hexbin()`: Hexagonal bin plot <br>


### Importing required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
```

### Sample Data

```python
#create data with timeseries. Using Numpy random for data and date_range for index.
print("Sample data DF1, DF2.")

df1 = pd.DataFrame(np.random.rand(1000,4), columns=list('ABCD'), index=pd.date_range('1/1/2015', periods=1000))
print("DF1 Data:(1000 X 4) \n", df1.head(5))

df2 = pd.DataFrame(np.random.rand(10,4), columns=list('ABCD'))
print("\n\nDF2 Data:(10 X 4) \n", df2.head(5))
```

<hr>

## 1) Line plot()

- `Line plot()` is useful to plot lines using DataFrame’s values as coordinates. This is default plot.

```python
print("1-EX1: The plot on Series or DataFrame is just a simple wrapper around plt.plot(): We have 2 syntaxes for same output.")

df2.plot(kind='line')
df2.plot.line()
```

```python
print("1-EX2: Subplots for each columns, if we apply lineplot to entire DF.")

df1.plot.line(subplots=True, linewidth=0.5)
```

```python
print("1-EX3: X vs Y line graph. X=Date , Y=Column A")

df1['A'].plot.line(figsize=(12,3), lw=1)
```

<hr>

## 2) Bar plot()

- `Bar plot()` will presents categorical data with rectangular bars with lengths proportional to the values.<br>
- `Bar plot()` shows comparisons among discrete categories. One axis of the plot shows the specific categories being compared, and the other axis represents a measured value.

```python
print("2-EX1: Plot for entire DF.")

df2.plot.bar()
```

```python
print("2-EX2: Plot for entire DF with Stacked one on one. DF2 have only 10 records. so it is easy.")

df2.plot.bar(stacked=True)
```

```python
print("2-EX3: Plot for entire DF with subplots.")

df2.plot.bar(subplots=True, figsize=(5,5))
```

```python
print("2-EX3: Plot for Single column in DF.")

df2['A'].plot.bar()
```

```python
print("2-EX4: Plot for X, Y in DF.")

df2.plot.bar(x='A', y='B')
```

## 4) Hist plot()

- `Hist plot()`: `Histograms`:  is a representation of the distribution of data. <br>
- `Hist plot()`: This function groups the values of all given Series in the DataFrame into bins and draws all bins. <br>

```python
print("4-EX1: For single columns A in DF1.")

df1['A'].plot.hist()
```

```python
print("4-EX2: For single columns A in DF1. with customized #of bins.")

df1['A'].plot.hist(bins=50)
```

<hr>

## 5) Box plot()

- `Box plot()` used for graphically groups of numerical data through their quartiles.  <br>
-  The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2).The position of the whiskers is set by default to 1.5*IQR (IQR = Q3 - Q1) from the edges of the box. <br> 

- Steps: <br>
> a) sort the entire data, <br>
> b) Find Median of entire data as Q2. #of elements/2. (0th element to n-1 element /2) <br>
> c) Find Median in first haf(0 the element to Q2) mark it Q1. Find Median in second haf(Q2 to n-1 element) mark it Q3. <br>
> d) IQR (interquartile range) = Q3-Q1 <br>
> e) Min = Q1-(1.5 X IQR), Max = Q3+(1.5 X IQR) <br>
> f) outof Min, Max treated as Outliers. <br>


```python
print("5-EX1: Box plot for entire DF")

df2.plot.box()
```

```python
print("5-EX2: Box plot for 1 column in DF")

df2['B'].plot.box()
```

```python
print("5-EX3: Best example.")
x = pd.Series([4, 17, 7, 14, 18, 12, 3, 16, 10, 4, 4, 11, 22, 44, 1])

x.plot.box(cmap='coolwarm')
```

```python
print("6-EX1: Use for entire DF")

df2.plot.kde()
```

```python
print("6-EX1: Use for 1 column in  DF")

df2['A'].plot.kde()
```

## 10) Scatter plot()

- `Scatter plot()`The coordinates of each point are defined by two dataframe columns and filled circles are used to represent each point.  <br>
- `Scatter plot()`is useful to see complex correlations between two variables <br>
- Points could be for instance natural 2D coordinates like longitude and latitude in a map or, in general, any pair of metrics that can be plotted against each other.

```python
print("10-EX1: X, Y values.")

df1.plot.scatter(x='A', y='B')
```

## 11) Hexbin plot()

- `Hexbin plot()` Generate a hexagonal binning plot of x versus y. <br>

```python
print("11-EX2: X vs Y")
df1.plot.hexbin(x='A', y='B', gridsize=15, cmap="viridis")
```
