# Pandas Build-in Data Visualization:

- In `Pandas` we can able to generate the visualizations wihtout using `Matplotlib`, `seaborn` libraries. <br>
- In the backend `Pandas` uses `Matplotlib` for data visualizations. <br>
- If we want to visualize some data in quick way(without worring too much about parameters) we use this Pandas Build-in stuff for visualization. <br>
- We are using below Plot() types. **Note:** df.plot(kind='hist') is same as df.plot.hist()
> 1) `df.plt.line()`: Line plot (default) <br>
> 2) `df.plt.bar()`: Vertical bar plot <br>
> 3) `df.plt.barh()`: Horizontal bar plot <br>
> 4) `df.plt.hist()`: Histogram <br>
> 5) `df.plt.box()`: Boxplot <br>
> 6) `df.plt.kde()`: Kernel Density Estimation plot <br>
> 7) `df.plt.density()`: it is same as ‘kde’ plot <br>
> 8) `df.plt.area()`: Area plot <br>
> 9) `df.plt.pie()`: Pie plot <br>
> 10) `df.plt.scatter()`: Scatter plot <br>
> 11) `df.plt.hexbin()`: Hexagonal bin plot <br>


### Importing required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

plt.style.use('fivethirtyeight')
```

### Sample Data

```python
#create data with timeseries. Using Numpy random for data and date_range for index.
print("Sample data DF1, DF2.")

df1 = pd.DataFrame(np.random.randn(1000,4), columns=list('ABCD'), index=pd.date_range('1/1/2015', periods=1000))
print("DF1 Data:(1000 X 4) \n", df1.head(5))

df2 = pd.DataFrame(np.random.randn(10,4), columns=list('ABCD'))
print("\n\nDF2 Data:(10 X 4) \n", df2.head(5))
```

## 1) Line plot()
- `Line plot()` is useful to plot lines using DataFrame’s values as coordinates. This is default plot.

```python
print("1-EX1: The plot on Series or DataFrame is just a simple wrapper around plt.plot(): We have 2 syntaxes for same output.")

df2.plot(kind='line')
df2.plot.line()
```

```python
print("1-EX2: Subplots for each columns, if we apply lineplot to entire DF.")

df1.plot.line(subplots=True)
```

```python
print("1-EX3: X vs Y line graph. X=Date , Y=Column A")

df1['A'].plot.line()
```
